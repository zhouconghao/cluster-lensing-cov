```python
import numpy as np
z_edges = np.array([0.2,0.35,0.5,0.65])
lam_edges = np.array([20, 45, 70, 120, 400])
```

```python
from astropy.cosmology import FlatLambdaCDM
from clens.util.parameters import CosmoParameters, NuisanceParameters
from clens.util.survey import Survey
from clens.lensing.cov_DeltaSigma import CovDeltaSigma
```

```python
co = CosmoParameters(h=0.701, OmegaDE=0.7352, OmegaM=0.2648, sigma8=0.8)  # CosmoDC2
nu = NuisanceParameters(
    sigma_lambda=0.18, lgM0=14.6258, alpha_M=1, lambda0=70
)  # Costanzi21 for now


survey_area = 1800.0
fsky = survey_area / 41253.0


for i in range(len(z_edges) - 1):
    for j in range(len(lam_edges) - 1):
        zh_min, zh_max = z_edges[i], z_edges[i+1]
        zs_mid = 2 * zh_max

        lambda_min, lambda_max = lam_edges[j], lam_edges[j+1]
        output_loc = "./des_y1_z_{}_{}_lam_{}_{}".format(zh_min,zh_max,lambda_min,lambda_max)

        su = Survey(
            top_hat=True, zs_min=zs_mid - 0.05, zs_max=zs_mid + 0.05, n_src_arcmin=6
)
        
        cp = DemoDESY1(
            co=co,
            nu=nu,
            su=su,
            zh_min=zh_min,
            zh_max=zh_max,
            lambda_min=lambda_min,
            lambda_max=lambda_max,
            rp_min_hiMpc=0.3,
            rp_max_hiMpc=5,
            n_rp=20,
            output_loc=output_loc,
        )
        
        cp.calc_cov_full(diag_only=False, fsky=fsky)
```
